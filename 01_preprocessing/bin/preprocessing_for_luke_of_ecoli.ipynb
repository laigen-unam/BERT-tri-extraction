```python
""" 
Get the start position and end position of a tagged string inside the original sentence.

:param pos: int, index of the tagged word in the sentence 
:param list_sent: list, the original sentence splitted into words  
:return: tuple, start idx and end idx of the tagged string
"""
def get_start_end(pos, list_sent):
    start = 0
    end = 0
    token_pos = 0
    for token in list_sent:
        return_list = []
        #print("token: {}".format(token))
        if pos == 0:
            start = 0
            end = start + len(token)
            return_list.append(start)
            return_list.append(end)
            return return_list 
        if token_pos == pos:
            # start += 1
            end = start + len(token)
            return_list.append(start)
            return_list.append(end)
            return return_list 
        start += len(token) + 1
        #print("start: {}".format(start))
        token_pos += 1
```

```python
span_list = []
regulator_list = []
regulated_list = []
for (sentence_o, sentence_t) in zip(data["SENTENCE"], data["sentence_tagged"]):
    print(sentence_o, sentence_t)
    list_sent_o = sentence_o.split()
    list_sent_t = sentence_t.split()
    regulator_pos = list_sent_t.index("@TF$")
    regulated_pos = list_sent_t.index("@Regulated$")
    print(regulated_pos)
    print(regulator_pos)
    # Get regulator TF idx inside the sentence 
    span_regulator= get_start_end(regulator_pos, list_sent_o)
    start_regulator = span_regulator[0]
    end_regulator = span_regulator[1]
    print("Start y end del regulador: {}".format(span_regulator))
    print("Regulador: {}".format(sentence_o[start_regulator:end_regulator]))
    # Get regulated idx inside the sentence
    span_regulated = get_start_end(regulated_pos, list_sent_o)
    start_regulated = span_regulated[0]
    end_regulated = span_regulated[1]
    print("Start y end del regulado: {}".format(span_regulated))
    print("Regulado: {}".format(sentence_o[start_regulated:end_regulated]))
    span_list.append([span_regulator, span_regulated])
    regulator_list.append(sentence_o[start_regulator:end_regulator])
    regulated_list.append(sentence_o[start_regulated:end_regulated])

```
